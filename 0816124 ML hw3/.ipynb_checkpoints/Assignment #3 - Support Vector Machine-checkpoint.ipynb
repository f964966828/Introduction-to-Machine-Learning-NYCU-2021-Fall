```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_table('svm_dataset.txt',header=None,sep=',')
df = shuffle(df)
display(df)
```

```python
x = df.iloc[:,:34]
y = df.iloc[:,34]

x_train, x_test = x[:math.floor(len(x)*0.7)], x[math.floor(len(x)*0.7):]
y_train, y_test = y[:math.floor(len(y)*0.7)], y[math.floor(len(y)*0.7):]
```

```python
clf=svm.SVC(kernel='linear')
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
kf = KFold(n_splits=5)

poly_best_acc = 0
poly_degrees = []
poly_gammas = []
poly_coefs = []
poly_accs = []
for degree in [0.5, 1, 5]:
    for gamma in [0.5, 1, 5]:
        for coef in [0.5, 1, 5]:
            # perform k-fold cross-validation
            clf=svm.SVC(kernel='poly', degree=degree, gamma=gamma, coef0=coef)

            total_y_kf_test = []
            total_y_kf_pred = []
            for train_index, test_index in kf.split(x_train):
                x_kf_train, x_kf_test = x_train.iloc[train_index], x_train.iloc[test_index]
                y_kf_train, y_kf_test = y_train.iloc[train_index], y_train.iloc[test_index]

                clf.fit(x_kf_train, y_kf_train)
                y_kf_pred = clf.predict(x_kf_test)

                total_y_kf_test += y_kf_test.tolist()
                total_y_kf_pred += y_kf_pred.tolist()

            acc = accuracy_score(total_y_kf_test, total_y_kf_pred)
            if acc > poly_best_acc:
                poly_best_acc = acc
                poly_best_degree = degree
                poly_best_gamma = gamma
                poly_best_coef = coef
    
            poly_degrees.append(degree)
            poly_gammas.append(gamma)
            poly_coefs.append(coef)
            poly_accs.append(acc)
        
df = pd.DataFrame()
df['degree'] = poly_degrees
df['gamma'] = poly_gammas
df['coef0'] = poly_coefs
df['accuracy'] = poly_accs
display(df)

print("Best parameter: degree = {}, gamma = {}, coef0 = {}".format(poly_best_degree, poly_best_gamma, poly_best_coef))
print("Best accuracy: {}".format(poly_best_acc))

clf=svm.SVC(kernel='poly', degree=poly_best_degree, gamma=poly_best_gamma, coef0=poly_best_coef)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
kf = KFold(n_splits=5)

rbf_best_acc = 0
rbf_gammas = []
rbf_accs = []
for gamma in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
    # perform k-fold cross-validation
    clf=svm.SVC(kernel='rbf', gamma=gamma)
    
    total_y_kf_test = []
    total_y_kf_pred = []
    for train_index, test_index in kf.split(x_train):
        x_kf_train, x_kf_test = x_train.iloc[train_index], x_train.iloc[test_index]
        y_kf_train, y_kf_test = y_train.iloc[train_index], y_train.iloc[test_index]
        
        clf.fit(x_kf_train, y_kf_train)
        y_kf_pred = clf.predict(x_kf_test)
        
        total_y_kf_test += y_kf_test.tolist()
        total_y_kf_pred += y_kf_pred.tolist()
        
    acc = accuracy_score(total_y_kf_test, total_y_kf_pred)
    if acc > rbf_best_acc:
        rbf_best_acc = acc
        rbf_best_gamma = gamma
        
    rbf_gammas.append(gamma)
    rbf_accs.append(acc)
        
df = pd.DataFrame()
df['gamma'] = rbf_gammas
df['accuracy'] = rbf_accs
display(df)

print("Best parameter: gamma = {}".format(rbf_best_gamma))
print("Best accuracy: {}".format(rbf_best_acc))

clf=svm.SVC(kernel='rbf', gamma=rbf_best_gamma)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
