```python
df = pd.read_csv('./data/train.csv')
ingredients = pd.read_csv('./data/ingredients.csv')['0'].tolist()
cuisines = pd.read_csv('./data/cuisines.csv')['0'].tolist()
df = shuffle(df, random_state=10111)
display(df)
```

```python
def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    torch.backends.cudnn.deterministic = True
# 设置随机数种子
setup_seed(20)

cuisines_map = dict()
for idx, cuisine in enumerate(cuisines):
    cuisines_map[cuisine] = idx
ingredients_map = dict()
for idx, ingredient in enumerate(ingredients):
    ingredients_map[ingredient] = idx
```

```python
label_count = [len(df[df['cuisine'] == cuisines_map[cuisine]]) for cuisine in cuisines]
labels = cuisines.copy()

label_count, labels = zip(*sorted(zip(label_count, labels)))
plt.figure()
plt.bar(labels, label_count, color='red', alpha =0.5)
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=90)
plt.xlabel('cuisine')
plt.ylabel('count')
plt.show()
```

```python
label_count = np.array([df[str(idx)].sum() for idx in range(len(ingredients))])
labels = np.array(ingredients)[np.argpartition(label_count, -30)[-30:]]
label_count = label_count[np.argpartition(label_count, -30)[-30:]]

label_count, labels = zip(*sorted(zip(label_count, labels)))
plt.figure()
plt.bar(labels, label_count, color='red', alpha =0.5)
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=90)
plt.xlabel('ingredient')
plt.ylabel('count')
plt.show()
```

```python
label_count = df['num_ingredient'].value_counts().array
labels = df['num_ingredient'].value_counts().index

plt.figure()
plt.bar(labels, label_count, color='red', alpha =0.5)
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=90)
plt.xlabel('number of ingredient')
plt.ylabel('count')
plt.show()
```

```python
class MyDataset(Dataset):
    def __init__(self, X, Y):
        self.X = X
        self.Y = Y
    def __len__(self):
        return len(self.X)
    def __getitem__(self, index):
        x = self.X[index]
        y = self.Y[index]
        return x, y
    
class DNN_model(nn.Module):
    def __init__(self):
        super(DNN_model, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(6214, 1024),  
            nn.ReLU(),
            nn.BatchNorm1d(1024),
            torch.nn.Dropout(0.7),
            nn.Linear(1024, 1024),  
            nn.ReLU(),
            nn.BatchNorm1d(1024),
            torch.nn.Dropout(0.7),
            nn.Linear(1024, 1024),
            nn.ReLU(),
            nn.BatchNorm1d(1024),
            torch.nn.Dropout(0.7), 
            nn.Linear(1024, 20)
        )

    def forward(self, x):
        return self.fc(x)

model = DNN_model().cuda()

summary(model, (6214, ))
```

```python
batch_size = 128
X = np.array(df.iloc[:, :-1]).astype(np.float32)
Y = np.array(df.iloc[:, -1])

train_x, val_x= X[:math.floor(len(X)*0.90)], X[math.floor(len(X)*0.90):]
train_y, val_y= Y[:math.floor(len(Y)*0.90)], Y[math.floor(len(Y)*0.90):]
train_set = MyDataset(train_x, train_y)
val_set = MyDataset(val_x, val_y)
train_loader = DataLoader(train_set, batch_size=batch_size)
val_loader = DataLoader(val_set, batch_size=batch_size)
```

```python
# model = DNN_model().cuda()
model = torch.load('./model/best_model1.pt')

loss = nn.CrossEntropyLoss()
# optimizer = torch.optim.Adam(model.parameters(), lr=1e-4, weight_decay=5e-3)
optimizer = torch.optim.SGD(model.parameters(), lr=5e-5, weight_decay=0, momentum=0.9)
num_epoch = 100

best_val_acc = 0

for epoch in range(num_epoch):
    epoch_start_time = time.time()
    train_acc = 0.0
    train_loss = 0.0
    val_acc = 0.0
    val_loss = 0.0

    model.train()
    for i, data in enumerate(train_loader):
        optimizer.zero_grad() 
        train_pred = model(data[0].cuda()) 
        batch_loss = loss(train_pred, data[1].cuda())
        batch_loss.backward() 
        optimizer.step()

        train_acc += np.sum(np.argmax(train_pred.cpu().data.numpy(), axis=1) == data[1].numpy())
        train_loss += batch_loss.item()
    
    model.eval()
    with torch.no_grad():
        for i, data in enumerate(val_loader):
            val_pred = model(data[0].cuda())
            batch_loss = loss(val_pred, data[1].cuda())

            val_acc += np.sum(np.argmax(val_pred.cpu().data.numpy(), axis=1) == data[1].numpy())
            val_loss += batch_loss.item()

        print('[%03d/%03d] %2.2f sec(s) Train Acc: %3.6f Loss: %3.6f | Val Acc: %3.6f loss: %3.6f' % \
            (epoch + 1, num_epoch, time.time()-epoch_start_time, \
             train_acc/train_set.__len__(), train_loss/train_set.__len__(), val_acc/val_set.__len__(), val_loss/val_set.__len__()))
        
    if best_val_acc < val_acc/val_set.__len__():
        best_val_acc = val_acc/val_set.__len__()
        torch.save(model, './model/model.pt')
    
    if best_val_acc > 0.8:
        optimizer = torch.optim.SGD(model.parameters(), lr=5e-5, weight_decay=0, momentum=0.9)
print("best_val_acc: {}".format(best_val_acc))
```

```python
model = torch.load('./model/best_model1.pt').cuda()
y_pred = model(torch.tensor(val_x).cuda())
y_pred = np.argmax(y_pred.cpu().data.numpy(), axis=1)
display(pd.DataFrame(confusion_matrix(val_y, y_pred)))
print(classification_report(val_y, y_pred))
```

```python
batch_size = 128
X = np.array(df.iloc[:, :-1]).astype(np.float32)
Y = np.array(df.iloc[:, -1])

train_x, val_x= X[:math.floor(len(X)*0.90)], X[math.floor(len(X)*0.90):]
train_y, val_y= Y[:math.floor(len(Y)*0.90)], Y[math.floor(len(Y)*0.90):]

aug_train_x = np.array(aug.iloc[:, :-1]).astype(np.float32)
aug_train_y = np.array(aug.iloc[:, -1])
train_x = np.concatenate((train_x, aug_train_x), axis=0)
train_y = np.concatenate((train_y, aug_train_y), axis=0)

train_set = MyDataset(train_x, train_y)
val_set = MyDataset(val_x, val_y)
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, drop_last=True)
val_loader = DataLoader(val_set, batch_size=batch_size, drop_last=True)
```

```python
# model = DNN_model().cuda()
model = torch.load('./model/best_model2.pt')

loss = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=5e-4)
# optimizer = torch.optim.SGD(model.parameters(), lr=5e-5, weight_decay=1e-3, momentum=0.99)
num_epoch = 100

best_val_acc = 0

for epoch in range(num_epoch):
    epoch_start_time = time.time()
    train_acc = 0.0
    train_loss = 0.0
    val_acc = 0.0
    val_loss = 0.0

    model.train()
    for i, data in enumerate(train_loader):
        optimizer.zero_grad() 
        train_pred = model(data[0].cuda()) 
        batch_loss = loss(train_pred, data[1].cuda())
        batch_loss.backward() 
        optimizer.step()

        train_acc += np.sum(np.argmax(train_pred.cpu().data.numpy(), axis=1) == data[1].numpy())
        train_loss += batch_loss.item()
    
    model.eval()
    
    with torch.no_grad():
        for i, data in enumerate(val_loader):
            val_pred = model(data[0].cuda())
            batch_loss = loss(val_pred, data[1].cuda())

            val_acc += np.sum(np.argmax(val_pred.cpu().data.numpy(), axis=1) == data[1].numpy())
            val_loss += batch_loss.item()

        print('[%03d/%03d] %2.2f sec(s) Train Acc: %3.6f Loss: %3.6f | Val Acc: %3.6f loss: %3.6f' % \
            (epoch + 1, num_epoch, time.time()-epoch_start_time, \
             train_acc/train_set.__len__(), train_loss/train_set.__len__(), val_acc/val_set.__len__(), val_loss/val_set.__len__()))
        
    if best_val_acc < val_acc/val_set.__len__():
        best_val_acc = val_acc/val_set.__len__()
        torch.save(model, './model/model2.pt')
        
    if best_val_acc > 0.795:
        # optimizer = torch.optim.SGD(model.parameters(), lr=5e-5, weight_decay=0, momentum=0.9)
        optimizer = torch.optim.SGD(model.parameters(), lr=5e-5, weight_decay=1e-4, momentum=0.99)
print("best_val_acc: {}".format(best_val_acc))
```
